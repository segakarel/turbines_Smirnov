## Задача 1:

Построить график зависимости термического КПД паротурбинного цикла без промежуточного перегрева пара при следующих параметрах пара: $P_0$ = 5, 10, 15, 20 MPa. Для каждого значения взять следующие значения температуры $t_0$ = 300, 350, 400, 450, 500 $^{\circ} $C, $p_k$ = 5 kPa. Принять давление за последней ступенью паровой турбины $p_2 = p_k$. Термический КПД цикла оценивать без учета подогрева воды в питательном насосе и регенеративной системе.тивной системе.

```python
import iapws
import matplotlib.pyplot as plt 
import numpy as np
from iapws import IAPWS97 as gas
MPa = 10 ** 6
kPa = 10 ** 3
unit = 1 / MPa
to_kelvin = lambda x: x + 273.15 if x else None


p_k = 5 * kPa
p_0 = {"P = 5 МPa":5 * MPa,"P = 10 МPa": 10 * MPa,"P = 15 МPa": 15* MPa,"P = 20 МPa": 20 * MPa}
t_0 = [300, 350, 400, 450, 500]
```

```python
def efficiency(p0: float, t0: float, pk: float) -> float:
    point_0 = gas(P = p0 * unit, T = t0)
    point_condenser_inlet = gas(P=pk * unit, s=point_0.s)
    useful_energy = point_0.h - point_condenser_inlet.h
    point_pump_outlet = gas(P= pk * unit, x=0)
    full_energy = point_0.h - point_pump_outlet.h
    efficiency = useful_energy / full_energy
    return efficiency

t_0 = [to_kelvin(t_0[i]) for i in range(len(t_0))]
```

```python
efficiency_dict = {}

for i in range(len(t_0)):
    efficiency_dict[str(t_0[i])] = [efficiency([i for i in p_0.values()][j], t_0[i], p_k) for j in range(4)]
```

```python
plt.figure(figsize=(10, 6))
plt.title('Зависимость $\eta$ от температуры', fontsize=16)
plt.xlabel('Температура (K)', fontsize=14)
plt.ylabel('$\eta$', fontsize=14)
plt.legend()  
plt.grid(True)
plt.plot(efficiency_dict.keys(),efficiency_dict.values(), marker='o')
plt.legend(p_0.keys())
plt.show()
```

# Задача 2

Построить график зависимости термического КПД паротурбинного цикла без промежуточного перегрева пара при следующих параметрах пара: $P_0$ = 5 MPa, $t_0 = 450 ^{\circ}$C, $p_k$ = 5, 10, 15, 20, 50 kPa. Принять давление за последней ступенью паровой турбины $p_2 = p_k$. Термический КПД цикла оценивать без учета подогрева воды в питательном насосе и регенеративной системе.тивной системе.

```python
# Начальные параметры:
P_0 = 5 * MPa
T_0 = to_kelvin(450)
P_k = {"P = 5 kPa":5 * kPa,"P = 10 kPa": 10 * kPa,"P = 15 kPa": 15 * kPa,"P = 20 kPa": 20 * kPa,"P = 50 kPa": 50 * kPa}
def get_key(d, value):
    for k, v in d.items():
        if v == value:
            return k
```

```python
            
point_0 = gas(P = P_0 * unit, T = T_0)
efficiency_dictt = {}

for i in P_k.values():
    j = get_key(P_k, i)
    efficiency_dictt[j] = efficiency_dictt.get(j, []) + [efficiency(P_0, T_0, i)]
```

```python
plt.figure(figsize=(10, 6))

plt.plot(P_k.values(),efficiency_dictt.values(), marker='o')

plt.title('Зависимость $\eta$ от давления $P_k$', fontsize=16)
plt.xlabel('Давление (Pa)', fontsize=14)
plt.ylabel('$\eta$', fontsize=14)
plt.legend()  
plt.grid(True)
plt.show()
```
